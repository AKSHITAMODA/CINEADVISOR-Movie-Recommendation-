```python
import pandas as pd
import numpy as np

df1 = pd.read_csv("tmdb_5000_credits.csv")
df2 = pd.read_csv("tmdb_5000_movies.csv")

df1.head()
```

```python
df2[1:3]
```

```python
df1.columns = ["id","title","cast","crew"]
df2 = df2.merge(df1 , on = "id")

df2[1:2]
```

```python
C = df2["vote_average"].mean()
print(C)
```

```python
m = df2["vote_count"].quantile(0.9)
print(m)
```

```python
q_movies = df2.copy().loc[df2["vote_count"] >= m]
print(q_movies)
```

v/(v+m))*R)+((m/(v+m))*C



1.   v = Vote count
2.   m = minimum vote to qualify
3.   R = Mean of the votes
4.   C = average voting



```python
def weighted_rating(x , m=m , C=C):
  v = x["vote_count"]
  R = x["vote_average"]
  return ((v/(v+m))*R)+((m/(v+m))*C)

q_movies["score"] = q_movies.apply(weighted_rating , axis=1)
print(q_movies["score"])
```

```python
q_movies = q_movies.sort_values("score",ascending=False)
q_movies[["original_title" , "vote_count" , "vote_average" , "score"]].head(10)
```

```python
import plotly.express as px

df = q_movies.head(10).sort_values("score" , ascending=True)

fig = px.bar(df , x = "score" , y = "original_title" , orientation = "h")
fig.show()
```

```python
def plot_top10_votes(df):
    fig = px.bar(
        df, x='original_title', y='vote_count',
        title='Top 10 Movies by Vote Count',
        labels={'original_title': 'Movie Title', 'vote_count': 'Vote Count'},
        text='vote_count',
        color_discrete_sequence=['green']
    )
    fig.update_layout(xaxis_tickangle=-45)
    fig.show()

top10_votes = df2.nlargest(10, 'vote_count')
print("Displaying Top 10 Movies by Vote Count...")
plot_top10_votes(top10_votes)
```

```python
def plot_top10_popularity(df):
    fig = px.bar(
        df, x='original_title', y='popularity',
        title='Top 10 Movies by Popularity',
        labels={'original_title': 'Movie Title', 'popularity': 'Popularity Score'},
        text='popularity',
        color_discrete_sequence=['purple']
    )
    fig.update_layout(xaxis_tickangle=-45)
    fig.show()
top10_popularity = df2.nlargest(10, 'popularity')
print("Displaying Top 10 Movies by Popularity...")
plot_top10_popularity(top10_popularity)
```

```python
def plot_top10_revenue(df):
    fig = px.bar(
        df, x='original_title', y='revenue',
        title='Top 10 Movies by Revenue',
        labels={'original_title': 'Movie Title', 'revenue': 'Revenue ($)'},
        text='revenue',
        color_discrete_sequence=['red']
    )
    fig.update_layout(xaxis_tickangle=-45)
    fig.show()
top10_revenue = df2.nlargest(10, 'revenue')

print("Displaying Top 10 Movies by Revenue...")
plot_top10_revenue(top10_revenue)
```

```python
from ast import literal_eval
#data parsing
features = ['cast', 'crew', 'keywords', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(literal_eval)

df2.dtypes
```

```python
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan

df2['director'] = df2['crew'].apply(get_director)
```

```python
def get_list(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        return names
    return []

features = ['cast', 'keywords', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(get_list)
```

This code makes the data easier to work with by transforming columns like cast, keywords, and genres into lists of names instead of lists of dictionaries. This simplification is useful for subsequent operations, like searching, filtering, or generating recommendations based on these features.
changes
[{'name': 'Leonardo DiCaprio'}, {'name': 'Kate Winslet'}]
to
['Leonardo DiCaprio', 'Kate Winslet']

```python
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''

features = ['cast', 'keywords', 'director', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(clean_data)
```

```python
def create_soup(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])
df2['soup'] = df2.apply(create_soup, axis=1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df2['soup'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
```

```python
df2 = df2.reset_index()
indices = pd.Series(df2.index, index=df2['original_title'])
```

```python
def get_recommendations(title, cosine_sim):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]
    return df2['original_title'].iloc[movie_indices]
```

```python
def filter_movies_by_runtime(df2, min_runtime=None, max_runtime=None):
    if min_runtime is not None:
        df2 = df2[df2['runtime'] >= min_runtime]

    if max_runtime is not None:
        df2 = df2[df2['runtime'] <= max_runtime]

    return df2['original_title']

time=int(input())
min_runtime = time-5
max_runtime = time+5
filtered_movies = filter_movies_by_runtime(df2, min_runtime=min_runtime, max_runtime=max_runtime)

for index,movie in enumerate(filtered_movies[0:10]):
  print(index+1,movie)
```

```python
def filter_movies_by_director(df, director_name):
    # Filter movies by the given director name
    df = df2[df2['director'] == director_name]
    return df['original_title']

director_name = input().lower()
director_name=director_name.replace(" ","")
filtered_movies = filter_movies_by_director(df2, director_name)

for index,movie in enumerate(filtered_movies[0:10]):
  print(index+1,movie)
```

```python
def filter_movies_by_language(df, language):
    language = language.lower().strip()
    df = df[df['original_language'].str.lower() == language]
    return df['original_title']

language_input = input("Enter language (e.g., 'en' for English, 'fr' for French, 'zh' for Chinese): ")
filtered_movies_by_language = filter_movies_by_language(df2, language_input)

print("Movies in the specified language:")
for index, movie in enumerate(filtered_movies_by_language[:10]):
    print(index+1,movie)
```

```python
def filter_movies_by_cast(df, cast_name):
    cast_name = cast_name.lower().replace(" ", "")
    df = df[df['cast'].apply(lambda x: cast_name in x)]
    return df['original_title']

cast_name = input("Enter cast name: ").lower().replace(" ", "")
filtered_movies_by_cast = filter_movies_by_cast(df2, cast_name)

print("Movies with the specified cast member:")
for index, movie in enumerate(filtered_movies_by_cast[:10]):
    print(index + 1, movie)

```

```python
# Define the apply_filters function to handle multiple filter criteria
def apply_filters(df, runtime=None, director=None, cast=None, language=None, genre=None):
    filtered_df = df.copy()  # Start with a copy of the original DataFrame

    if runtime:
        min_runtime, max_runtime = runtime - 5, runtime + 5
        filtered_df = filtered_df[(filtered_df['runtime'] >= min_runtime) & (filtered_df['runtime'] <= max_runtime)]

    if director:
        director = director.lower().replace(" ", "")
        filtered_df = filtered_df[filtered_df['director'] == director]

    # Apply cast filter
    if cast:
        cast = cast.lower().replace(" ", "")
        filtered_df = filtered_df[filtered_df['cast'].apply(lambda x: cast in x if isinstance(x, list) else False)]

    # Apply language filter
    if language:
        language = language.lower().strip()
        filtered_df = filtered_df[filtered_df['original_language'].str.lower() == language]

    # Apply genre filter
    if genre:
        genre = genre.lower().replace(" ", "")
        filtered_df = filtered_df[filtered_df['genres'].apply(lambda x: genre in x if isinstance(x, list) else False)]
    if filtered_df.empty:
        return pd.Series([])

    return filtered_df['original_title']

# Prompt user inputs for each filter criteria
runtime = int(input("Enter desired runtime (in minutes): "))
director_name = input("Enter director name (leave blank if not filtering by director): ").strip()
cast_name = input("Enter cast member's name (leave blank if not filtering by cast): ").strip()
language = input("Enter language (e.g., 'en' for English, leave blank if not filtering by language): ").strip()
genre = input("Enter genre (leave blank if not filtering by genre): ").strip()

# Apply filters based on user input
filtered_movies = apply_filters(df2, runtime=runtime, director=director_name, cast=cast_name, language=language, genre=genre)

# Display the filtered movies
print("Filtered Movies:")
for index, movie in enumerate(filtered_movies[:10]):  # Display the top 10 filtered results
    print(f"{index + 1}. {movie}")
```
